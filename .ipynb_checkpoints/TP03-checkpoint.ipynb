L'objectif de ce TP est de cerner les avantages de l'informatique quantique à travers l'exemple de l'algorithme de Deutsch-Jozsa. 

## Deutsch-Jozsa en informatique classique

Pour évaluer si une fonction prenant n paramètres est constante ou équilibrée, il nous faudrait en informatique classique (n/2)+1 éxécutions de cette dernière. En effet, si l'intégralité de ces éxéutions sont égales, alors la fonction est constante, sinon elle est équilibrée.

## Deutsch-Jozsa en informatique quantique

En informatique quantique, nous pouvons définir des oracles, prenant n qubits en entrée et évaluer si ces derniers sont équilibrés ou constants d'une seule éxécution.

![blip bloup](oracle_IQ01.png)

```python
from qiskit import *
import matplotlib
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
```

## Oracle constant

```python
#Création d’un circuit quantique instance de QuantumCircuit
circ = QuantumCircuit(4,3)

circ.h(0)
circ.h(1)
circ.h(2)
circ.barrier()
# Annule la précédente porte Hadamard
circ.h(0)
circ.h(1)
circ.h(2)

circ.barrier()

circ.measure(0, 0)
circ.measure(1, 1)
circ.measure(2, 2)

circ.draw('mpl')
```

## Oracle équilibré

```python

balanced_circ = QuantumCircuit(4,3)
balanced_circ.h(0)
balanced_circ.h(1)
balanced_circ.h(2)
balanced_circ.x(3)
balanced_circ.h(3)

balanced_circ.barrier()
# Opération CNOT 
balanced_circ.cx(0, 3)
balanced_circ.cx(1, 3)
balanced_circ.cx(2, 3)
# Les qubits sont à présent superposés
balanced_circ.barrier()
balanced_circ.h(0)
balanced_circ.h(1)
balanced_circ.h(2)
balanced_circ.barrier()
# Mesure des 3 qubits initiaux. Ils devraient tous être égaux si le circuit est équilibré.
balanced_circ.measure(0, 0)
balanced_circ.measure(1, 1)
balanced_circ.measure(2, 2)

balanced_circ.draw('mpl')
```

```python
# Evaluation de l'oracle
simulator = Aer.get_backend('aer_simulator')

compiled_circuit = transpile(balanced_circ, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()
counts = result.get_counts(compiled_circuit)
plot_histogram(counts, title='résultats')
```

# Pour n qubit

```python
nbQbit= int(input("Nb de Qbit d'entrée"))
balanced_circ = QuantumCircuit(nbQbit+1,nbQbit)
#Ajout de portes H pour superposition dans l'état initial 01
for i in range (nbQbit):
    balanced_circ.h(i)
balanced_circ.x(nbQbit)
# Permet de mesurer si le système est équilibré ou non
balanced_circ.h(nbQbit)
balanced_circ.barrier()
for i in range (nbQbit):
    balanced_circ.cx(i, nbQbit)
balanced_circ.barrier()
# Inversion de la porte H initiale
for i in range (nbQbit):
    balanced_circ.h(i)
balanced_circ.barrier()
# Mesures
for i in range (nbQbit):
    balanced_circ.measure(i, i)


balanced_circ.draw('mpl')
```

```python
# Evaluation de l'oracle
simulator = Aer.get_backend('aer_simulator')
compiled_circuit = transpile(balanced_circ, simulator)
job = simulator.run(compiled_circuit, shots=1000)
result = job.result()
counts = result.get_counts(compiled_circuit)
plot_histogram(counts, title='résultats')
```

Nous savons donc avec certitude que notre oracle est équilibré.
